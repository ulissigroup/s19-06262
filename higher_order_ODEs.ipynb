# In-class example 1

Solve y''=y

```octave
function yprime = diff_eq(t,y)
    yprime = [y(2);y(1)];
end

trange=linspace(0,10);

sol=ode45(@diff_eq, trange,[1 2]);
plot(sol.x,sol.y(1,:))
legend('y')
```

# In-class example 2

Solve y'''+2y''-y'-2y=0

```octave
function yprime = diff_eq(t,y)
    yprime = [0 1 0;
                0 0 1;
                2 1 -2]*y;
end

trange=linspace(0,10);

sol=ode45(@diff_eq, trange,[1;-1;1]);
plot(sol.x,sol.y(1,:))
legend('y')
```

# In-class example 3

Solve y''''+t^2y+y''/y'''+y^2y'+cos(t)=0

```octave
function yprime = diff_eq(t,y)
    yprime = [y(2); #y'=y'
            y(3); #y''=y''
            y(4); #y'''=y'''
            -cos(t)-y(1)^2*y(2)-y(3)/y(4)-t^2*y(4)];
end

trange=linspace(0,0.2);

sol=ode45(@diff_eq, trange,[1;1;1;1]);
plot(sol.x,sol.y(1,:))
legend('y')
```

# In-class example 4

```octave
#Systems of higher order diff eq
function yprime = diff_eq(t,y)
    yprime = [y(2); #z' = z'
                y(2)+y(3)+cos(t); 
                y(4); #w' = w'
                y(1)+cos(t)-t];
end

trange=linspace(0,1);
sol=ode45(@diff_eq, trange,[1;1;1;1]);
hold all

#Plot z and w, not z' and w'
plot(sol.x,sol.y(1,:))
plot(sol.x,sol.y(3,:))
legend('z','w')
```
