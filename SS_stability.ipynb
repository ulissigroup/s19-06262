# In-class example 1

3 steady state solutions at y=[0,0], y=[3,9], y=[-2,4]

```octave
function yprime = diff_eq(t,y)
    yprime = [y(1)*y(2)-y(2)-6*y(1);
            -y(1)^2+y(2)];
end

#Solve and plot at first steady state
ode45(@diff_eq, [0 10],[0 0]);
legend('y1','y2')
hold all 

#Solve and plot at second steady state
ode45(@diff_eq, [0 10],[3 9]);
legend('y1','y2')

#Solve and plot at third steady state
ode45(@diff_eq, [0 10],[-2 4]);
legend('y1','y2')
```

## Stability of first solution

```octave
y=[0;0];
eig([y(2)-6 y(1)-1;-2*y(1) 1])
```

One eigenvalue is positive, one is negative

```octave
#Solve and plot at first steady state
ode45(@diff_eq, [0 10],[0.,0.]);
legend('y1','y2')
ode45(@diff_eq, [0 10],[.01,.01]);
legend('y1','y2')
```

## Stability of second solution

```octave
y=[3;9]
eig([y(2)-6 y(1)-1;-2*y(1) 1])
```

eigenvalues are complex and real part is positive. Solution near this SS will look something like e^2t*sin(x). Unstable!

```octave
function yprime = diff_eq(t,y)
    yprime = [y(1)*y(2)-y(2)-6*y(1);
            -y(1)^2+y(2)];
end

ode45(@diff_eq, [0 10],[2.9 8.9])
```

Unstable! The system moves away from the steady state to a new steady state!

## Stability of third SS solution

```octave
y=[-2,4]
eig([y(2)-6 y(1)-1;-2*y(1) 1])
```

```octave
function yprime = diff_eq(t,y)
    yprime = [y(1)*y(2)-y(2)-6*y(1);
            -y(1)^2+y(2)];
end

ode45(@diff_eq, [0 10],[-2.5 4.5])
ode45(@diff_eq, [0 10],[1 1])
```

Very stable to perturbations

# Rabbit/wolf example

```octave
% Check the eigenvalues of our jacobian
eig([0 -0.2*0.4;
    2.5 0])
```

```octave
#First, check that this is actually a steady state
```

```octave
function yprime = rabbit_wolf(t,y)
    yprime = [y(1)-0.2*y(1)*y(2);
                0.5*y(1)*y(2)-0.2*y(2)];
end

ode45(@rabbit_wolf,[0,5],[0.4 5])
```

```octave
function yprime = rabbit_wolf(t,y)
    yprime = [y(1)-0.2*y(1)*y(2);
                0.5*y(1)*y(2)-0.2*y(2)];
end

ode45(@rabbit_wolf,[0,20],[0.5 5.5])
```

The linear approximation is actually really good near the steady state! It looks like just sin/cos functions
